20. **reserved_room_type** : code of room type reserved
21. **assigned_room_type** : Code of room type assigned
22. **booking_changes** : Number of changes/amendments made to the booking

29. **required_car_parking_space** : Number of car parking spaces required by the customer

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the current dataset and assigning to df
df_copy = df.copy()

# Drop column "company" Because this column having many missing values
df_copy.drop('company', axis=1, inplace=True)

# impute missing values in "agent", "country" and "children"
df_copy['children'].fillna(0, inplace=True)
df_copy['agent'].fillna(0, inplace=True)
df_copy['country'].fillna('Other', inplace=True)
````

````python
# Replace 0 with not_cancelled and 1 with Cancelled in is_canceled column.
df_copy['is_canceled'] = df_copy['is_canceled'].replace([0,1], ['Not Cancelled', 'Cancelled'])

# Same here we will do for is_repeated_guest col.
df_copy['is_repeated_guest'] = df_copy['is_repeated_guest'].replace([0,1], ['Not Repeated', 'Repeated'])
````

````python
# Total stay in nights
df_copy['total_stay_nights'] = df_copy['stays_in_weekend_nights'] + df_copy['stays_in_week_nights']
df_copy['total_stay_nights']
````

````python
# Now We have to create a col for revenue using total_stay * adr
df_copy['revenue'] = df_copy['total_stay_nights'] * df_copy['adr']
df_copy['revenue']
````

````python
# Now we will check over all revenue by hotel type
hotel_wise_total_revenue = df_copy.groupby('hotel')['revenue'].sum()
hotel_wise_total_revenue
````

````python
df_copy[['hotel', 'revenue']]
````

````python
# Also, for information, we will add a column with total guest for each booking
df_copy['total_guests'] = df_copy['adults'] + df_copy['children'] + df_copy['babies']
df_copy['total_guests'].sum()
````

#### Chart - 8 - Price variation over the year

````python
# Here we will groupby "arrival_date_month" with "adr"
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
p = df_copy.groupby(['hotel','arrival_date_month'])['adr'].mean().reset_index()
p['arrial_date_month'] = pd.Categorical(p['arrival_date_month'],  categories=months, ordered=True)
p.sort_values(by='arrival_date_month', inplace=True)
````

````python
# Plotting
plt.figure(figsize=(15,8))
sns.lineplot(data=p, x='arrival_date_month', y='adr', hue='hotel')
plt.xlabel('Month')
plt.ylabel('Average Daily Price')
plt.title('Average Daily Rate By Month')
p = plt.xticks(rotation=30)
plt.show()
````

##### 1. Why did you pick the specific chart?

A line chart is a great choice for visualizing trends over time. In this scenario, we are plotting the average daily rate (ADR) for different hotel types across months, making a line chart an effective way to show how ADR fluctuates throughout the year and compare the patterns between hotel types.
